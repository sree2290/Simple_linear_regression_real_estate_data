# Simple Linear Regression Using traditional Statictical Methods

### Importing Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("real_estate_price_size.csv")
data
```

### Discriptive Statistics

```python
data.describe()
```

```python
y = data["price"]
X1 = data["size"]
```

### Scatter Plot for Two numarical Variables

```python
plt.scatter(X1,y);
```

### Statistics of the Model using StatsModels.api


```python
x = sm.add_constant(X1)
results = sm.OLS(y,x).fit()
results.summary()
```

### Drew a Linear line to conncet all dots

```python
plt.scatter(X1,y)
yhat =   223.1787* X1 + 1.019e+05	
fig = plt.plot(X1,yhat, c = "red", lw = 2)
```
